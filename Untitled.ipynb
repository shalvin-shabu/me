# median



```python
def median(n,a):
   
    for i in range(0,n):
        for j in range(i+1,n):
            if(a[i]>a[j]):
                temp=a[i]
                a[i]=a[j]
                a[j]=temp
    if(n%2!=0):
        return(a[n//2])
    else:
        return((a[n//2]+a[(n+1)//2])/2)
n=int(input("enter the limit "))
print("enter the elements")
a=[int(input())for i in range(n)]
sortlist=median(n,a)

print("median=",sortlist)
```

# mode


```python
def mode(a):
    result=[]
    dict1=dict.fromkeys(a,0)
    for i in a:
        if i in dict1:
            dict1[i]+=1
    for key in dict1:
        if dict1[key]==max(dict1.values()):
            result.append(key)
    return result
n=int(input("enter the limit "))
print("enter the elements")
a=[int(input())for i in range(n)]
print("mode=",mode(a))
```

# password checking

```python
import re
def check_password_validity(password):
    error=[]
    if(len(password)<8):
        error.append("atleast 8 characters")
    if not re.search("[a-z]",password):
        error.append("atleast one lowercase")
    if not re.search("[A-Z]",password):
        error.append("atleast one uppercase")
    if not re.search("[0-9]",password):
        error.append("atleast one numbercase")
    if not re.search("[!@#$%^&*]",password):
        error.append("atleast one special character")
    if error:
        print("try again".join(error))
    else:
        print("password is valid")
        
password=input("enter the password")
check_password_validity(password)
```

```python
def timeofday(n):
    if n in range(1,4):
        return 'Late Night'
    elif n in range(4,7):
        return 'Early Morning'
    elif n in range(7,12):
        return 'Morning'
    elif n in range(12,15):
        return 'After Noon'
    elif n in range(15,18):
        return 'Evening'
    elif n in range(18,25):
        return 'Night'
df['timeofday']=df['hour'].apply(lambda x:timeofday(x))
```

```python
def monthtoseasons(n):
    if n in[9,10,11]:
        return 'spring'
    elif n in[12,1,2]:
        return 'summer'
    elif n in[3,4,5]:
        return 'autumn'
    elif n in [6,7,8]:
        return 'winter'
df['seasons']=df['month'].apply(lambda x:monthtoseasons(x))
```
